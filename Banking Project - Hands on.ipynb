```python
from abc import ABCMeta, abstractmethod
import numpy as np

class Account(metaclass = ABCMeta):
    
    @abstractmethod
    def create_account():
        return 0
    @abstractmethod
    def authenticate():
        return 0
    @abstractmethod
    def withdraw():
        return 0
    @abstractmethod
    def deposit():
        return 0
    @abstractmethod
    def display_balance():
        return 0
    
class Saving_account(Account):
    
    def __init__(self):
        # [key][0] => name ; [key][1] => balance
        self.saving_accounts = {}
    
    def create_account(self, name, initial_deposit):
        print()
        self.account_id = np.random.randint(10000, 99999) 
        self.saving_accounts[self.account_id] = [name, initial_deposit]
        print("Account has been created successful. Your account number is ", self.account_id)
        print()

    def authenticate(self, name, account_id):
        print()
        
        try:
            if (self.saving_accounts[account_id][0] == name):
                print()
                print('Account has been authenticated successfully')
                self.account_id = account_id
            
                return True
        except:
                print()    
                print("Authentication failed")
            
                return False
       
            
    def withdraw(self, withdraw_amount):
        if self.saving_accounts[account_id][1] < withdraw_amount:
            print()
            print('The balance is insufficient')
           
        else:
            self.saving_accounts[account_id][1] -= withdraw_amount
            print()
            print('The withdrawl is complete successfully , Available Balance:', end='')
            self.display_balance()
             
    def deposit(self, deposit_amount):
        if deposit_amount >= 0 :
            self.saving_accounts[account_id][1] += deposit_amount
            print()
            print('The deposit is complete successfully , Available Balance:', end = '')
            self.display_balance()
             
        else:
            print()
            print('The deposit amount must be greater than 0')
            
    def display_balance(self):
        print()
        print('The Availabel Balance:', self.saving_accounts[account_id][1])
        
        
```

```python

# initialized instance
saving_account = Saving_account()

while True:

    print('Enter 1 to create a new account')
    print('Enter 2 to access an existing account')
    print('Enter 3 to exit')
    
    user_choice = int(input())
    
    if user_choice is 1:
        print('Enter your name:')
        name = input()
        print('Enter your Initial Deposit:')
        initial_deposit = int(input())
        saving_account.create_account(name, initial_deposit)
        
    elif user_choice is 2:
        
        print('Enter your name:')
        name = input()
        print('Enter your account number:')
        account_id = int(input())
        
        authenticated_status = saving_account.authenticate(name, account_id)
        
        if authenticated_status is True:
           
            while True:
                print('Enter 1 to withdraw')
                print('Enter 2 to deposit')
                print('Enter 3 to display available balance')
                print('Enter 4 to go back to the previous manu')
                user_choice = int(input())

                if user_choice is 1:
                    print('Please enter the withdrawl amount:')
                    withdrawl_amount = int(input())
                    saving_account.withdraw(withdrawl_amount)

                elif user_choice is 2:
                    print('Please enter the deposit amount:')
                    deposit_amount = int(input())
                    saving_account.deposit(deposit_amount)


                elif user_choice is 3:

                    saving_account.display_balance()

                elif user_choice is 4:
                    break
                
           # authenticated_status is not True , will print authentication failed automatically
           # has been built in function
       
            
        
    elif user_choice is 3:
        break
```
